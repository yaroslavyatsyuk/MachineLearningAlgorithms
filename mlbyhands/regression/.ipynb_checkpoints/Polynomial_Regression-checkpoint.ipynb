```python
class PolinomialRegression:
    '''
        Newton Method for Polynomial Regression
    '''
    CONVERGENCE_THRESHOLD = 1.5e-8    
    
    def __init__(self, degree=1):
        self.degree = degree
        
    def _to_polinomial_features(self, X, degree):
        if degree <= 1:
            return X
        else:
            iterator = 0
            poly_features = np.empty((len(X), X.shape[1] * degree))
            print(poly_features.shape)
            for i in range(X.shape[1]):
                for power in range(1, degree + 1):
                    poly_features[:,iterator] = X[:,i]**power
                    iterator += 1

            return poly_features
        
    def _gradient(self, X, y, weights):
        '''Gradient vector from RSS'''
        m = len(X)
        return -2 * X.T @ (y - X @ weights)

    def _hessian(self, X):
        '''Hessian matrix from RSS'''
        m = len(X)
        return 2 * X.T @ X

    def _cost_fuction(self, X, y, theta):
        '''Mean Square Error'''
        m = y.shape[0]
        return np.sum((y - X @ theta)**2) / m  

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        
        X = self._to_polinomial_features(X, self.degree)
        
        # Add intercept as first column
        X = np.insert(X, 0, 1, axis=1)
        
        self.weights = np.zeros(X.shape[1])
        
        theta = self.weights
        
        prev_cost = self._cost_fuction(X, y, theta)
        theta = self.weights - (np.linalg.inv(self._hessian(X)) @ self._gradient(X, y, theta))
        cost = self._cost_fuction(X, y, theta)
        while(abs(cost - prev_cost) > self.CONVERGENCE_THRESHOLD):
            theta = theta - (np.linalg.inv(self._hessian(X)) @ self._gradient(X, y, theta))
            prev_cost = cost
            cost = self._cost_fuction(X, y, theta)
        
        self.weights = theta
            
    def predict(self, X):
        X = to_polinomial_features(X, self.degree)
        X = np.insert(np.array(X), 0, 1, axis=1)
        return X @ self.weights
```

```python
def plot_regression_fit(X, y, feature_name, model, title):
    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(8,7))
    plt.title(title)
    plt.scatter(X[feature_name], y, alpha=0.3)
    plt.plot(X[feature_name], model.predict(X), '-', color="purple")
```
